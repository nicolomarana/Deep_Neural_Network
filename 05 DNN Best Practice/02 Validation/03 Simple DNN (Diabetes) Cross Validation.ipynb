```
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras import backend
from sklearn.model_selection import StratifiedKFold
```

```
# Fix seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```
# load pima indians dataset
filename='https://gist.githubusercontent.com/nicolomarana/728b2063e19ff4c6d56790a041693767/raw/f450082c7e2917545fda1a5bda10c4faad647ade/diabetes'
dataset = numpy.loadtxt(filename, delimiter=',', skiprows=1)
```

```
# Dividiamo in input(X) e output(Y)
X = dataset[:,0:8]
y = dataset[:, 8]
```

```
# define 10-fold cross validation test harness
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
cvscores = []
```

```
import numpy as np
from sklearn.model_selection import StratifiedKFold


skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
skf.get_n_splits(X, y)

print(skf)

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = Sequential()
    model.add(Dense(12, input_dim=8, activation="relu", kernel_initializer="uniform"))
    model.add(Dense(8, activation="relu", kernel_initializer="uniform"))
    model.add(Dense(1, activation="sigmoid", kernel_initializer="uniform"))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # Fit the model
    model.fit(X_train, y_train, epochs=150, batch_size=10, verbose=0)
    # evaluate the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)

print("%.2f%% (+/- %.2f%%)" % (numpy.mean(cvscores), numpy.std(cvscores)))
```
