```
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras import backend
from sklearn.model_selection import train_test_split 
```

```
# Fix seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```
# load pima indians dataset
filename='https://gist.githubusercontent.com/nicolomarana/728b2063e19ff4c6d56790a041693767/raw/f450082c7e2917545fda1a5bda10c4faad647ade/diabetes'
dataset = numpy.loadtxt(filename, delimiter=',', skiprows=1)
```

```
# Dividiamo in input(X) e output(Y)
X = dataset[:,0:8]
Y = dataset[:, 8]
```

```
# split into 67% for train and 33% for test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=seed)
```

```
# crea il modello
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer= 'uniform', activation='relu'))
model.add(Dense(8,  kernel_initializer='uniform', activation='relu'))
model.add(Dense(1,  kernel_initializer='uniform', activation='sigmoid'))
```

```
# Model information
model.summary()
```

```
# compila il modello
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
# fit il modello
model.fit(X_train, y_train, validation_data=(X_test,y_test), nb_epoch=150, batch_size=10)
```
